```python
path = "/content/taxi.csv"
data = pd.read_csv(path)
```

```python
print(data.head())
print(data.tail())
len(data)
```

```python
data_x=data.iloc[:, 0:-1].values
data_y=data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split(data_x, data_y, test_size=0.3, random_state=0)
```

```python
x_train, x_test, y_train, y_test  = train_test_split(data_x, data_y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_pred=reg.predict(x_test)
y_pred # Prints predicted value
```

```python
y_test # Shows Actual Value
```

```python
print(reg.coef_) # Prints coefficient values ie. b1,b2,b3,b4
print(reg.coef_[0])
```

```python
print(reg.intercept_) # Prints intercept values i.e b0
```

```python
# y= b0 + b1x1+ b2x2 + b3x3 + b4x4

y_0= reg.intercept_+reg.coef_[0]*x_test[0][0]+reg.coef_[1]*x_test[0][1]+reg.coef_[2]*x_test[0][2]+reg.coef_[3]*x_test[0][3]
y_1= reg.intercept_+reg.coef_[0]*x_test[1][0]+reg.coef_[1]*x_test[1][1]+reg.coef_[2]*x_test[1][2]+reg.coef_[2]*x_test[1][3]
```
