```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression  #Library for model built
```

```python
#pwd
```

```python
data = pd.read_csv('/content/Salary_Data.csv')# Loading data set
data.head(5)
```

```python
# real_x=data.iloc[:,0].values # iloc for index location[row, column]
# real_y=data.iloc[:,1].values
# print(real_y)
# real_x=real_x.reshape(-1,1)
# real_y=real_y.reshape(-1,1)

real_x=data.iloc[:,:-1].values # iloc for index location[row, column]
print(real_x)
real_y=data.iloc[:,1].values
print(real_y)
```

```python
training_x,testing_x,training_y,testing_y = train_test_split(real_x,real_y,test_size=.2,random_state=0) # training and testing for supervise learning for this we have to include sklearn library
# .3 means 30% for testing and remaining 70% for training and 30% for testing

Lin=LinearRegression() # for building a model
Lin.fit(training_x,training_y)# fit function for applying train the algo.training complete through fit method
```

```python
pred_y=Lin.predict(testing_x)# for prediction completion
testing_y[3] # random value for testing third column(actual values)
```

```python
pred_y[3] # predicted values. see below 63282.41 is approx 60 no difference from values shown above i.e. prediction is highly accurate
```

```python
plt.scatter(training_x,training_y,color='red') # plot training data
plt.plot(training_x,Lin.predict(training_x),color='blue') # Lin.predict(training_x) is prediction data for training. plt .plot()  used for how to plot Regression line
plt.title("salary and Experience training plot")

plt.xlabel("Ëxp")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(testing_x,testing_y,color='red')
plt.plot(testing_x,Lin.predict(testing_x),color='blue')
plt.title("salary and Experience testing plot")

plt.xlabel("Ëxp")
plt.ylabel("Salary")
plt.show() # give best fitting line through ordinary least square method
```
